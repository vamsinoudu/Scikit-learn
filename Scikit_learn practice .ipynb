```python
x=df.drop('Exited',axis=1)
```

```python
y=df['Exited']
```

# OneHotEncoding:

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categ_feature=["Geography","Gender"]

onehot=OneHotEncoder()

transformer=ColumnTransformer([("onehot",onehot,categ_feature)],remainder="passthrough")

transformed_x=transformer.fit_transform(x)

transformed_x
```

```python
df.head(10)
```

```python
pd.DataFrame(transformed_x)
```

# Training and Testing the Data:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,r2_score
rnd=RandomForestClassifier()
```

```python
rnd.fit(x_train,y_train)
```

```python
pred=rnd.predict(x_test)
```

```python
print(accuracy_score(y_test,pred))
```
